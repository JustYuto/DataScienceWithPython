## 4..モデル構築と検証
x -> 入力値

t -> 目標値

```python
x = df.drop('y', axis=1)
t = df['y']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# モデルの宣言(インスタンス化)
model = LinearRegression()
```

```python
#　学習
model.fit(x, t)
```

```python
#　検証
model.score(x, t)
```

## 5. 訓練データと検証データ

例）受験勉強　10年分の過去問を購入

悪いケース

- 10年分で勉強（学習）
- 10年分で実力テスト（検証）

良いケース
- ５年分でランダムに勉強（学習）→訓練データ
- ５年分で答えを見ずに実力テスト（検証）→検証データ

```python
from sklearn.model_selection import train_test_split
```

```python
#分割
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.4, random_state=1) #randome_stateは乱数のシードの固定(再現性の確保)
```

```python
len(x_train), len(x_test)
```

```python
len(t_train), len(t_test)
```

```python
#モデルの学習
model.fit(x_train, t_train)
```

```python
#検証（訓練データ）
model.score(x_train, t_train)
```

```python
#検証（訓練データ）
model.score(x_test, t_test)
```
